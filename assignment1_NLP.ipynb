```python
import nltk
from nltk.tokenize import (
    word_tokenize,
    WhitespaceTokenizer,
    TreebankWordTokenizer,
    TweetTokenizer,
    MWETokenizer
)
from nltk.stem import PorterStemmer, SnowballStemmer
from nltk.stem import WordNetLemmatizer

# Download required NLTK data
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
text = "NLTK is amazing! I'm learning NLP, tokenization, stemming, and lemmatization ðŸ˜Š #NLP"
print("Original Text:\n", text)
```

```python
whitespace_tokenizer = WhitespaceTokenizer()
tokens_whitespace = whitespace_tokenizer.tokenize(text)

print("Whitespace Tokenization:")
print(tokens_whitespace)
```

```python
tokens_punctuation = word_tokenize(text)

print("Punctuation-Based Tokenization:")
print(tokens_punctuation)
```

```python
treebank_tokenizer = TreebankWordTokenizer()
tokens_treebank = treebank_tokenizer.tokenize(text)

print("Treebank Tokenization:")
print(tokens_treebank)
```

```python
tweet_tokenizer = TweetTokenizer()
tokens_tweet = tweet_tokenizer.tokenize(text)

print("Tweet Tokenization:")
print(tokens_tweet)
```

```python
mwe = MWETokenizer([('machine', 'learning'), ('natural', 'language', 'processing')], separator='_')

sample_text = "I am learning machine learning and natural language processing"
tokens_mwe = mwe.tokenize(sample_text.split())

print("MWE Tokenization:")
print(tokens_mwe)
```

```python
porter = PorterStemmer()
porter_stems = [porter.stem(word) for word in tokens_punctuation]

print("Porter Stemmer Output:")
print(porter_stems)
```

```python
snowball = SnowballStemmer("english")
snowball_stems = [snowball.stem(word) for word in tokens_punctuation]

print("Snowball Stemmer Output:")
print(snowball_stems)
```

```python
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in tokens_punctuation]

print("Lemmatization Output:")
print(lemmatized_words)
```
